# Question 1

```python
import pandas as pd
import numpy as np
```

bring the dataframe

```python
url_html='https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&direction=next&oldid=942655364'

df_html = pd.read_html(url_html)

df=df_html[0]

df.head(10)
```

Drop 'Not assigned' from Borough

```python
indexNames = df[ df['Borough'] == 'Not assigned' ].index
 
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
```

```python
df.head(10)
```

if Neighbourhood is 'Not assigned' than replace with Borough

```python
df['Neighbourhood'].replace(np.nan, df['Borough'])
df.head(10)
```

```python
df.drop_duplicates(subset='Postcode', keep='first', inplace=True)
df.head(10)
```

```python
df.shape
```
